let us wrangle some more

```python
# replace U with NaN
df.replace('U', np.nan, inplace=True)
# cast object data type to integer data type
df = df.astype(int, errors='ignore')
# call describe again
df.describe()
```

# Lazy Evaluation with Dask (and PySpark)

Operations are stacked behind the scenes and an optimized transformation is calculated.

The result is not calculated immediately, as Pandas would do with eager evaluation.

When an operation requires the results, then the result is calculated.

This allows for processing of big data:

terabytes and petabytes

noticable decrease in performance

memory limitations

To obtain results one can force an evaluation with the compute method

```python
ddf.std().compute()
```

instead of using read_csv one can also create a dask.dataframe with from_pandas

npartitions argument defines the number of partitions to divide the data into, so one can work across these partitions

```python
ddf = dd.from_pandas(df, npartitions = 2)
```

```python
ddf
```

with Dask one can also utilize methods in a similar fashion as with Pandas

```python
result = ddf['Lives saved'].sum() - ddf['Lives lost, total'].sum()
result.compute()
# difference between lives lost and lives saved
```

overall more lives have been saved than lost

```python
result_verify = ddf['Lives lost before notificationc,d'].sum() - ddf['Lives lost after notificatione'].sum()
result_verify.compute()
# difference between lives lost before and after notification, and resulting response
```

more lives have been lost before notification overall

Conclusions

Choosing a Framework
* Pandas or NumPy vs PySpark or Dask: based on performance and size of data
* Pandas or NumPy: based on nature of data (types, dimensions, complexity of operations, desire to perform analysis and create visualizations)
* PySpark or Dask: based on nature of the enterprise
* PySpark: mature, support, documentation, may already be in use
* Dask: Python native, may use full capabilities of Pandas and NumPy, may utilize if starting a big data project and desire pipeline to be all in Python
